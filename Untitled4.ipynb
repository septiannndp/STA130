1. **Accuracy** measures the proportion of true results (both true positives and true negatives) in the population.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
2. **Sensitivity** measures the proportion of actual positives that are correctly identified.

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$

3. **Specificity** measures the proportion of actual negatives that are correctly identified.

$$\text{Specificity} = \frac{TN}{TN + FP}$$

4. **Precision** measures the proportion of positive identifications that were actually correct.

$$\text{Precision} = \frac{TP}{TP + FP}$$

### Answer:
1. Real-World Application: Customer Satisfaction Surveys
- Accuracy is most helpful when the categories, like "Satisfied" and "Unsatisfied," are evenly balanced, and mistakes in either direction (false positives or false negatives) are equally important. For example, in a survey collecting customer feedback, the main goal would be to correctly identify both satisfied and unsatisfied customers, with neither type of mistake being more serious than the other.

2. Real-World Application: Medical Diagnosis for Critical Diseases (e.g., Cancer Detection)
- Sensitivity is important when it’s critical to find all positive cases and avoid missing any. For example, in cancer screening, missing a diagnosis could delay treatment and lead to serious problems. High sensitivity makes sure almost all real cases are found, even if it means there might be a few false alarms.

3. Real-World Application: Spam Email Filtering
- Specificity matters when it’s important to correctly identify negative cases and avoid false positives. For example, in spam filtering, marking real emails as spam by mistake can cause inconvenience or missed messages. High specificity helps ensure that most real emails aren’t filtered out incorrectly.
4. Real-World Application: Fraud Detection in Financial Transactions
- Precision is important when false positives are a problem. In fraud detection, having too many false alarms can waste investigators' time or upset honest customers flagged by mistake. High precision means the flagged transactions are more likely to be actual fraud, making the process more effective and reliable.


### 3. Explore the amazon books dataset, seen previously at the start of the semester, providing some initital standard *exploratory data analysis (EDA)* and data summarization after pre-processing the dataset to meet the requirements below<br>

### Answer:

````python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import accuracy_score, recall_score, make_scorer
import graphviz as gv

url = "https://raw.githubusercontent.com/pointOfive/STA130_F23/main/Data/amazonbooks.csv"
ab = pd.read_csv(url, encoding="ISO-8859-1")

# Remove specified columns
ab_reduced_noNaN = ab.drop(columns=['Weight_oz', 'Width', 'Height'])

# Drop rows with NaN entries
ab_reduced_noNaN = ab_reduced_noNaN.dropna()

# Set the data types of the specified columns
ab_reduced_noNaN['Pub year'] = ab_reduced_noNaN['Pub year'].astype(int)
ab_reduced_noNaN['NumPages'] = ab_reduced_noNaN['NumPages'].astype(int)
ab_reduced_noNaN['Hard_or_Paper'] = ab_reduced_noNaN['Hard_or_Paper'].astype('category')
````

### 5. Repeat the previous problem but this time visualize the *classification decision tree* based on the following specifications below; then explain generally how predictions are made for the *clf2* model<br>

1. `X = ab_reduced_noNaN[['NumPages', 'Thick', 'List Price']]`
2. `max_depth` set to `4`

### Answer:

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

# Prepare the training data (using only List Price as the feature)
X1_train = ab_reduced_noNaN_train[['NumPages', 'Thick', 'List Price']]
y_train = pd.get_dummies(ab_reduced_noNaN_train['Hard_or_Paper'])['H']

# Initialize the DecisionTreeClassifier with max_depth=2
clf2 = DecisionTreeClassifier(max_depth=4, random_state=42)

# Train the classifier on the training data
clf2.fit(X1_train, y_train)

# Plot the tree to explain predictions based on List Price
plt.figure(figsize=(12, 8))
tree.plot_tree(clf2, feature_names=['NumPages', 'Thick', 'List Price'], class_names=['Paper', 'Hard'], filled=True)
plt.show()
````

Predictions are made by starting at the root of the decision tree and using decision rules based on features like NumPages, Thickness, and List Price to split the data at each node. The tree selects the feature and threshold that best separates the data to improve classification accuracy. As the model moves down the branches, it applies the rules based on the input values, making decisions at each node until it reaches a leaf node. Each leaf node holds a prediction, which shows the probability or count of books being hardcover or paperback for the data that reaches it. The final prediction is the class with the majority in the leaf node. For example, if a book’s List Price is above a certain threshold and NumPages falls within a specific range, the tree might predict it as a hardcover, whereas other ranges could lead to a paperback classification.

````python
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

# Prepare the test data (using only List Price as the feature for clf)
X_test = ab_reduced_noNaN_test[['List Price']]
y_test = pd.get_dummies(ab_reduced_noNaN_test['Hard_or_Paper'])['H']

# Predictions for clf (model 1) on test data
y_pred_clf = clf.predict(X_test)

# Confusion matrix for clf
conf_matrix_clf = confusion_matrix(y_test, y_pred_clf)
print("Confusion Matrix for clf (Test Data):")
print(conf_matrix_clf)

# Sensitivity (Recall for "Hard" class)
sensitivity_clf = recall_score(y_test, y_pred_clf, pos_label=1)
# Specificity (Recall for "Paper" class)
specificity_clf = recall_score(y_test, y_pred_clf, pos_label=0)
# Accuracy
accuracy_clf = accuracy_score(y_test, y_pred_clf)

# Display metrics for clf
print("\nMetrics for clf (Test Data):")
print(f"Sensitivity: {sensitivity_clf}")
print(f"Specificity: {specificity_clf}")
print(f"Accuracy: {accuracy_clf}")

# Plot confusion matrix for clf (Test Data)
plt.figure(figsize=(6, 6))
ConfusionMatrixDisplay(conf_matrix_clf, display_labels=['Paper', 'Hard']).plot(cmap='Blues')
plt.title('Confusion Matrix for clf (Test Data)')
plt.show()

# Prepare the test data (using NumPages, Thick, List Price as the features for clf2)
X1_test = ab_reduced_noNaN_test[['NumPages', 'Thick', 'List Price']]
y1_test = pd.get_dummies(ab_reduced_noNaN_test['Hard_or_Paper'])['H']

# Predictions for clf2 (model 2) on test data
y_pred_clf2 = clf2.predict(X1_test)

# Confusion matrix for clf2
conf_matrix_clf2 = confusion_matrix(y1_test, y_pred_clf2)
print("\nConfusion Matrix for clf2 (Test Data):")
print(conf_matrix_clf2)

# Sensitivity (Recall for "Hard" class)
sensitivity_clf2 = recall_score(y1_test, y_pred_clf2, pos_label=1)
# Specificity (Recall for "Paper" class)
specificity_clf2 = recall_score(y1_test, y_pred_clf2, pos_label=0)
# Accuracy
accuracy_clf2 = accuracy_score(y1_test, y_pred_clf2)

# Display metrics for clf2
print("\nMetrics for clf2 (Test Data):")
print(f"Sensitivity: {sensitivity_clf2}")
print(f"Specificity: {specificity_clf2}")
print(f"Accuracy: {accuracy_clf2}")

# Plot confusion matrix for clf2 (Test Data)
plt.figure(figsize=(6, 6))
ConfusionMatrixDisplay(conf_matrix_clf2, display_labels=['Paper', 'Hard']).plot(cmap='Blues')
plt.title('Confusion Matrix for clf2 (Test Data)')
plt.show()
````

### 9. Describe the differences of interpreting coefficients in linear model regression versus feature importances in decision trees in two to three sentences<br>

### Answer:
In linear regression models, the coefficients show how much each predictor variable affects the predicted outcome, with their size indicating how much the dependent variable changes when the predictor increases by one unit, assuming other variables stay the same. In decision trees, feature importance shows how much each feature helps reduce uncertainty or impurity (like Gini impurity) at each decision point, with higher values meaning the feature contributes more to accurate predictions. Unlike linear models, decision trees can capture complex, non-linear relationships, so feature importances reflect the overall impact of a feature in the tree’s structure, rather than just an additive effect.
